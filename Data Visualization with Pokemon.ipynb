## Section 1: Comparing Generations

To get started, I wanted to compare two generations of Pokemon, Generation 1 and Generation 6. By using the "Generation" column of the dataset, I can separate the two generations with the variables "genOne" and "genSix".

```python
genOne = data[data.Generation == 1]
genSix = data[data.Generation == 6]
```

```python
genOne
```

```python
genSix
```

Now, I'm using plt.plot to plot all of the Pokemon by their Id numbers (x-axis) and their Total (y-axis), which are the products of the Pokemon's base values.

```python
plt.plot(genOne.Id, genOne.Total)
plt.plot(genSix.Id - 649, genSix.Total)
plt.legend(['Gen 1', 'Gen 6'])
plt.xlabel('Id Number (from original game)')
plt.ylabel('Total Base Value')
plt.show()
```

From the chart shown above, we can conclude that:

. The total base values gradually increase the higher the id number is, indicating that the higher the id number, the stronger the pokemon

. There are about half as many pokemon created from the sixth generation than the first one
